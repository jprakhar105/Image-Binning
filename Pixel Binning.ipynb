```python
import cv2 as cv
import numpy as np
from numpy import asarray
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = "ClockTower.jpg"
image_display = Image.open(image_path)
plt.imshow(image_display)
```

```python
image_original = cv.imread(image_path)
array_original = asarray(image_original)
original_dimensions = array_original.shape
print(original_dimensions)
new_dimensions = (int(original_dimensions[0]/2), int(original_dimensions[1]/2), 3)
print(new_dimensions)
print(type(array_original[0][0][0]))
```

```python
array_binned = np.zeros(new_dimensions, dtype = np.uint8)
for i in range(0, new_dimensions[0]):
    for j in range(0, new_dimensions[1]):
        start_row, end_row = i * 2, (i + 1) * 2
        start_col, end_col = j * 2, (j + 1) * 2
        average_color = np.mean(image_original[start_row:end_row, start_col:end_col, :], axis=(0, 1))
        array_binned[i, j, :] = np.round(average_color).astype(np.uint8) 
```

```python
binned_image = Image.fromarray(array_binned, 'RGB')
plt.imshow(binned_image)
```
